Problem:

A population of U surgeons recieves N notifications a day. You are given the data in a simple CSV file containing a list of pairs (surgeon, notifications_per_day).

Please write a program that implements an algorithm to sort the surgeons by "notifications per day". However, let's say you have limited RAM, so you are not allowed to work on more than 3 surgeons at a time. For the sake of parsing simplicity, you are allowed to load the whole CSV in RAM, but the sorting cannot work on more than 3 surgeons at a time. 

```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv("./testdata_exercise2.csv", header = None)
col_names2 = ['surgeon', 'notifications_per_day']
df2.columns = col_names2
df2.head(10)
```

```python
df2.sort_values(by='notifications_per_day', ascending=False)
```

```python
# This is the standard bubblesort algorithm that sorts an array

def bubbleSort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(0, n - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
    return nums
```

```python
def bubbleSortDataframe(df):
    n = len(df)
    for i in range(n):
        for j in range(0, n - i - 1):
            if df.iloc[j]['notifications_per_day'] > df.iloc[j + 1]['notifications_per_day']:
                df.iloc[j], df.iloc[j + 1] = df.iloc[j + 1], df.iloc[j]

    return df
```

```python
# Turning the not_per_day column into an array 
nums = np.array(df2['notifications_per_day'])
```

```python
# Show the unordered array
print(nums)
```

```python
# Sorted array
bubbleSort(nums)
```

```python
# Sorting the df using bubbleSortDataframe function 
bubbleSortDataframe(df2)
```
